### Duplicate IDs

```python
# Discover and delete duplicate IDs
dupe_mask = id_counts == 2
dupe_mask[0:5]
```

```python
id_counts.index[0:5]
```

```python
# Identify duplicate numbers by using index as a parameter and turning them into a list
dupe_ids = id_counts.index[dupe_mask]
dupe_ids = list(dupe_ids)
len(dupe_ids)
```

```python
df.loc[df['ID'].isin(dupe_ids[0:3]),:].head(10)
```

```python
# Generate dataframe with Boolean values taking into account the rows where all values are zero
df_zero_mask = df == 0
df_zero_mask.head()
```

```python
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis = 1)
feature_zero_mask
```

```python
sum(feature_zero_mask)
```

```python
# Eliminating rows where all values are zeros

df_clean_1 = df.loc[~feature_zero_mask, :].copy()
df_clean_1
```

```python
# Comparing the size of the dataframe with the number of unique values in the "ID" columnprint(

print(
    df_clean_1['ID'].nunique(),
    df_clean_1.shape
)    
```

```python
# Informações sobre as colunas e o tipo de dado em cada uma delas 
df_clean_1.info()
```

```python
df_clean_1['PAY_1'].head(5)
```

```python
# Limpeza das linhas 'Not available'
valid_pay_1_mask = df_clean_1['PAY_1'] != 'Not available'
valid_pay_1_mask[:5]
```

```python
sum(valid_pay_1_mask)
```

```python
# Mesclando os dados da máscara  booleana com o DF principal
df_clean_2 = df_clean_1.loc[valid_pay_1_mask, :].copy()
df_clean_2.head()
```

```python
# Confirmando se o 'Not available' realmente saiu da contagem das linhas
df_clean_2['PAY_1'].value_counts()
```

```python
# Alterar o tipo de dado de uma coluna de 'object' para 'int' 
df_clean_2['PAY_1'] = df_clean_2['PAY_1'].astype('int64')
df_clean_2[['PAY_1', 'PAY_2']].info()
```
